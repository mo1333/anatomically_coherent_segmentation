```python
import glob
import os
import numpy as np
from PIL import Image
```

```python
path = "data_test/REFUGE2/Validation/Disc_Masks/"
new_path = "data_test/REFUGE2/Validation/New_Disc_Masks/"
glob_path = path + "*.png"

if not os.path.exists(new_path):
    os.makedirs(new_path)

count = 0

for img in glob.glob(glob_path):
    print(img)
    image = np.array(Image.open(img))
    new_image = np.transpose(np.stack([image < 120, image < 240]), (1, 2, 0))
    new_image = 255 * new_image.astype(np.uint8)
    new_image = Image.fromarray(new_image)
    new_image.save(new_path + img.split("\\")[-1])
```
